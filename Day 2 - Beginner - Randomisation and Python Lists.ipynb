# Code Challenge 1: Heads or Tails

You are going to write a virtual coin toss program. It will randomly tell the user "Heads" or "Tails". 

**Important**, the first letter should be capitalised and spelt exactly like in the example e.g. Heads, not heads.

There are many ways of doing this. But to practice what we learnt in the last lesson, you should generate a random number, either 0 or 1. Then use that number to print out Heads or Tails.

e.g.
1 means Heads
0 means Tails 

# Example Output

```

```

or

```
Tails
```



````python
import random

heads_tails = random.randint(0,1)
if heads_tails == 1:
    print('Heads')
else:
    print('Tails')
````

# Code Challenge 2: Oscar Nominations  - Oscar for best Actor goes to?

You are going to write a program which will select a random name from a list of names. The person selected will get oscar. 

**Important**: You are not allowed to use the `choice()` function.

**Line 8** splits the string `names_string` into individual names and puts them inside a **List** called `names`. For this to work, you must enter all the names as name followed by comma then space. e.g. name, name, name

```

```
Note: notice that there is a space between the comma and the next name. 
# Example Output



# Hint

1. You might need the help of the `len()` function.   


````python
# Split string method
names_string = input("And the Nomminee for best actor in lead roles are. ")
names = names_string.split(", ")
print(names)
oscar = random.randint(0,len(names))
if oscar == 0:
    print('And the Oscar goes to: ',names[0])
elif oscar == 1:
    print('And the Oscar goes to: ',names[1])
elif oscar == 2:
    print('And the Oscar goes to: ',names[2])
elif oscar == 3:
    print('And the Oscar goes to: ',names[3])
elif oscar == 4:
    print('And the Oscar goes to: ',names[4])
elif oscar == 5:
    print('And the Oscar goes to: ',names[5])
else:
    print('And the Oscar goes to: ',names[6])
````
